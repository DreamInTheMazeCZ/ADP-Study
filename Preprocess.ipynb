```python
# Q56. 파일 로드 및 행렬 개수 출력

card = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv', index_col=0)
card.shape
```

```python
card.head()
```

```python
# Q57. Income_Category 컬럼을 map 함수를 이용해 다음과 같이 변경하여 newIncome 컬럼에 매핑
# Unknown : N
# Less than $40K : a
# $40K - $60K : b
# $60K - $80K : c
# $80K - $120K : d
# $120K +’ : e

lst = sorted(card.Income_Category.unique(), reverse=True)
lst = lst[:2] + lst[2:-1][::-1] + lst[-1:]
mapper = ['N', 'a', 'b', 'c', 'd', 'e']

card['newIncome'] = card.Income_Category.map({l:m for l, m in zip(lst, mapper)})
card
```

```python
# Q58. Income_Category 컬럼을 apply 함수를 이용해 다음과 같이 변경하여 newIncome 컬럼에 매핑
dic = {'Unknown' : 'N',
       'Less than $40K' : 'a',
       '$40K - $60K' : 'b',
       '$60K - $80K' : 'c',
       '$80K - $120K' : 'd',
       '$120K +' : 'e'}

card['newIncome'] = card.Income_Category.apply(lambda x: dic[x])
card.newIncome.head()
```

```python
# Q59. Customer_Age의 값을 구간화한 AgeState 컬럼 생성 및 각 구간의 빈도수 출력 (0~9 : 0, 10~19 :10, 20~29 : 20, ...)
import math

card['AgeState'] = card.Customer_Age.apply(lambda x: math.floor(x / 10) * 10)
card.AgeState.value_counts().sort_index()
```

```python
# Q60. Education_Level 컬럼 중 Graduate가 포함되는 값은 1, 그렇지 않은 경우에는
# 0으로 변경하여 newEduLevel 컬럼 생성 및 빈도수 출력

card['newEduLevel'] = card.Education_Level.str.contains('Graduate')
card['newEduLevel'] = card['newEduLevel'].astype('int64')
card.newEduLevel.value_counts()
```

```python
# Q61. Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 컬럼 생성 및 빈도수 출력

card['newLimit'] = card.Credit_Limit.map(lambda x: 1 if x >= 4500 else 0)
card.newLimit.value_counts()
```

```python
# Q62. Marital_Status가 Married이고, Card_Category가 Platinum인 경우 1,
# 그 외의 경우에는 모두 0으로 하는 newState 컬럼 생성 및 빈도수 출력

card['newState'] = (card['Marital_Status'] == 'Married') & (card['Card_Category'] == 'Platinum')
card['newState'] = card['newState'].astype('int64')
card.newState.value_counts()
```

```python
# Q63. Gender가 M인 경우 male, F인 경우 female로 값 변경 및 빈도수 출력

gen = {'M':'male', 'F':'female'}
card['Gender'] = card.Gender.map(lambda x: gen[x])
card.Gender.value_counts()
```
